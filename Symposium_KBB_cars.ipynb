```python
Cars_data=[]
    
for i in range(len(cars)):
    main_page_url=cars[i]
    resp = requests.get(main_page_url)
    tagtree = html.fromstring(resp.content)
    
    title=''
    price=''
    Mileage=''
    Body_Style=''
    Exterior_Color=''
    Interior_Color=''
    Fuel_Economy=''
    Engine=''
    Fuel_Type=''
    Transmission=''
    Drive_Type=''
    Doors=''
    expert_rating=''
    consumer_rating=''
    
    #########Title###########################
    title_xp1=  '//h1[@class="primary-vehicle-title title-two "]/text()'
    title = tagtree.xpath(title_xp1)
    #########Price###########################
    price_xp1=  '//span[@class="price js-price"]/text()'
    price = tagtree.xpath(price_xp1)
    #################################
    Vehicle_xp1=  '//div[@class="details-list"]/ul/li/text()'
    vehicle = tagtree.xpath(Vehicle_xp1)

    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Mileage':
            mileage=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Body Style':
            Body_Style=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Exterior Color':
            Exterior_Color=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Interior Color':
            Interior_Color=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Engine':
            Engine=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Fuel Type':
            Fuel_Type=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Fuel Economy':
            Fuel_Economy=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Transmission':
            Transmission=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Drive Type':
            Drive_Type=vehicle[i].split(':')[1]
    for i in range(len(vehicle)):
        if vehicle[i].split(':')[0]=='Doors':
            Doors=vehicle[i].split(':')[1]
    
    ##########################################
    consumer_rating_xp1=  '//a[@class="ratings-reviews-link js-consumer-review"]/p/span/text()'
    consumer_rating_str = tagtree.xpath(consumer_rating_xp1)
    expert_rating_xp1=  '//a[@class="ratings-reviews-link js-expert-review"]/p/span/text()'
    expert_rating_str = tagtree.xpath(expert_rating_xp1)
    
    try:
        if str(consumer_rating_str[0]) != '':
            consumer_rating=str(consumer_rating_str[0])
        else:
            consumer_rating=''
    except IndexError:
            consumer_rating = ''
    
    try:
        if str(expert_rating_str[0]) != '':
            expert_rating=str(expert_rating_str[0])
        else:
            expert_rating=''
    except IndexError:
        consumer_rating = ''   
        
    Cars_data.append([main_page_url,title,price,mileage,Body_Style,Exterior_Color,Interior_Color,Fuel_Economy,Engine,Fuel_Type,Transmission,Drive_Type,Doors,consumer_rating,expert_rating])
```

```python
Cars_data_df=pd.DataFrame(Cars_data)
```

```python
pd.set_option('display.max_colwidth', -1)
Cars_data_df.columns=['Kbb_url','title','price','mileage','Body_Style','Exterior_Color','Interior_Color','Fuel_Economy','Engine','Fuel_Type','Transmission','Drive_Type','Doors','consumer_rating','expert_rating']
```

```python
Cars_data_df.to_csv('KBB_cars.csv',index=False)
```
